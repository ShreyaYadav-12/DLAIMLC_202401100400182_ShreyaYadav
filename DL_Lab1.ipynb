```python
import tensorflow as tf
import numpy as np

# step1:convert data to NumPy arrays
x=np.array([1,2,3,4], dtype=float)
y=np.array([2,4,6,8], dtype=float)

# step2:define model using Input layer (recommended)
model=tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),       #preferred over input_shape in Dense
    tf.keras.layers.Dense(1)
])

# step3:compile the model
model.compile(optimizer='sgd',loss='mean_squared_error')

# step4:train the model
model.fit(x,y,epochs=10)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# step1:convert lists to NumPy arrays
x=np.array([1,2,3,4], dtype=float)
y=np.array([2,4,6,8], dtype=float)

# step2:define model
model=Sequential([
    Dense(1, input_shape=(1,))   #this works, or you can use tf.keras.Input(shape=(1,))
])

# step3:compile the model
model.compile(optimizer='sgd',loss='mse')

# step4:train the model
model.fit(x,y,epochs=10)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# define a simple linear model
model=nn.Linear(1,1)

# loss and optimizer
criterion=nn.MSELoss()
optimizer=optim.SGD(model.parameters(), lr=0.01)

# dummy data
x=torch.tensor([[1.],[2.],[3.],[4.]])
y=torch.tensor([[2.],[4.],[6.],[8.]])

# training loop
for epoch in range(10):
  optimizer.zero_grad()
  outputs=model(x)
  loss=criterion(outputs,y)
  loss.backward()
  optimizer.step()
  print(f"Epoch {epoch+1}, Loss: {loss.item()}")
```
